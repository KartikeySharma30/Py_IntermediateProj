```python
                                                    ''' SLOT MACHINE '''
# How the Slot Machine Works ?
# So this project is about matching the coloumns from the no. of selected row which has randomly generated elements 
#     |
#     --> if Matched -> Get Reward associated with the element
#     |
#     --> if not Matched -> No Reward


import random

# Global Constants : So that we can use them anywhere in our program
# Constant values that we are not gona change its a Globle Constant 
MAX_LINES = 3   # We can also input this from the user but for this project we kept it as constants
MAX_BET = 100
MIN_BET = 1

ROWS = 3
COLS = 3

symbol_count = {    # These counts determins the no. of item in a single Reel  
    "A": 2,
    "B": 4,
    "C": 6,
    "D": 8
}

symbol_value = {    # Reward Associated with different different elements   
    "A": 5,
    "B": 4,
    "C": 3,
    "D": 2
}


# 7) Here the we check the total no. of winnings and incrementing its value to with the value associated with the symbols * bet
def check_winnings(columns,lines,bet,values):
    winning = 0
    winning_lines=[]
    for line in range(lines):
        symbol = columns[0][line]
        for column in columns :
            symbol_to_check = column[line]
            if symbol != symbol_to_check:
                break
        else:
            winning += values[symbol] * bet
            winning_lines.append(line + 1)
    return winning,winning_lines



# 5)
# Here we are first creating a list 
# all_symbols - in which we are storing all the symbols corresponding to the no. of times of their count 
# all-symbols = [ A,A,B,B,B,B,C,C,C,C,C,C,D,D,D,D,D,D,D,D]
# Then Creating two another list columns = [] and column = [] : where column would be nested under columns 
# So we are iterating over rows=3
#        and creating a list column =[] at each iteration 
#        and copying the list all_symbols into current_symbols
#        and iterating over cols=3
#            choosing random elements from the current_symbols list
#            removing the choosed element from the current_symbols so that avoid chances of repeatance
#            appending choosed value to column []
#        there after appending that row to columns [[],[],....]
# at last returning columns
def get_slot_machine_spin(rows,cols,symbols):
    # Here we will be creating a list from which we would be randomly selecting values and taking them out
    all_symbols = []
    for symbol,symbol_count in symbols.items():
        for _ in range(symbol_count):   # "_" refers to anonymus variable in python -> when you dont actually need the iteration value
            all_symbols.append(symbol)  # then it is used 

    columns = []
    for _ in range(rows):
        column = []
        current_symbols = all_symbols[:] # here we are copying the list so that at each iteration the element dont get remove from the original list
        # We are copying it in this way and not in current_symbols = all_symbols cos if we do the they both will point to the same data points in 
        # memory so if we tend to change the element even if we dont want the other will become the same automatically
        for _ in range(cols):
            value = random.choice(current_symbols)
            current_symbols.remove(value)
            column.append(value)
        columns.append(column)

    return columns


# 6)
# here we are Transposing the columns list that we formed
def print_slot_machine(columns): 
    for row in range(len(columns[0])): # as row and column are equal therefore there will be equal no of element in the list
        for i,column in enumerate(columns): # Enumerate give the index value
            if i !=len(columns[0])-1:
                print(column[row],end=" | ")
            else:
                print(column[row],end="")
        print()



# 2)
#So from here the user enter the desired amount to deposit in a variable -> <amount> and place bet from that 
# amount * <no. of row selected> : which gets deducted from the original deposited amount
# deposit() -> takes input from the user(in string Format) then from that verify it isdigit or not 
#                                                                                      |
#                                                                                      --> check is > 0 or not
# at last returns the amount
def deposit():
    while True:
        amount = input("What would you like to deposit ? $ ")
        if amount.isdigit():
            amount=int(amount)
            if amount > 0:
                break
            else:
                print("Please Enter the no. Greater than 0")

        else:
            print("Number is Required in this field!")
    return amount


# 3) 
# So Here we are taking input from the user and verifying it-> for the no. of lines we want the bet from top -> 1 to bottom -> 3
# So you can bet on :
#                   - row 1
#                   - row 1 and row 2
#                   - row 1 , row 2 and row 3 
# We can also customise it for choosing a particular rows as well
def get_number_of_lines():
    while True:
        line = input("Enter the Number of lines to bet on (1-" + str(MAX_LINES)+")? ")  # here str(MAX_LINES) -> returns the value in string
        if line.isdigit():                                                              # So that we can concatinate it with the whole str
            line=int(line)
            if 1<=line<=MAX_LINES : 
                break
            else:
                print("Please Enter the valid no. of lines between the range")

        else:
            print("Number is Required in this field!")
    return line



# 4)
# So here we are taking the amount to be betted on the selected lines as input
def get_bet():
    while True:
        amount = input("What would you like to bet on Each line? $ ")
        if amount.isdigit():
            amount=int(amount)
            if MIN_BET<=amount<=MAX_BET :
                break
            else:
                print(f"The Amount must be between ${MIN_BET} - ${MAX_BET}")
                
        else:
            print("Number is Required in this field!")
            
    return amount



# 1.2)
def spin(balance):
    lines = get_number_of_lines()
    while True :
        bet = get_bet()
        total_bet = bet * lines
        if total_bet > balance :
            print(f"You Do not have enough amount to bet that amount, your current balance is :${balance}")
        else:
            break
    print(f"You are betting ${bet} on {lines} lines. Total bet is equal to :${total_bet}")
    
    

    slots = get_slot_machine_spin(ROWS,COLS,symbol_count)
    print(slots)
    print_slot_machine(slots)
    winnings, winning_lines = check_winnings(slots,lines,bet,symbol_value)
    print(f"You won : {winnings} ")
    print(f"You won on lines :",*winning_lines)
    print("winnings - total_bet :",winnings - total_bet)
    return winnings - total_bet
# 1) From here and 1.2 all the all the main functions will be called to execute the tasks
def main():
    balance = deposit()
    while True:
        print(f"Current Balance is :${balance}")
        ans = input("Press Enter to Spin (q to quit)")
        if ans == 'q':
            break
        balance +=spin(balance)
        if balance == 0:
            break
    print(f"You left with :${balance}")
main()
```
